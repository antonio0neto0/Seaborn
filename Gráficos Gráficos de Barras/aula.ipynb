Tipos de Gráficos: Gráficos de Barras

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

produtos = ['Mochila', 'Fones de Ouvido', 'Relógio']
precos = [150, 200, 50]
anos = [2010,2020,2030]

produtos = pd.DataFrame({'Produtos':produtos,'Preços':precos,'Ano':anos})

# Personalização padrão
sns.set_theme(style='darkgrid')
plt.figure(figsize=(5,5),dpi=150)

# Barplot(gráfico de barra)
cores1 = ['#32a873', '#323ea8', '#a032a8']
cores2 = ['red','yellow','blue']
# sns.barplot(x='Produtos',y='Preços',data=produtos,hue='Ano',palette='bright')
# Gráfico de barra Vertical
sns.barplot(x='Produtos',y='Preços',data=produtos,hue='Ano',palette=cores2)
plt.show()
# Gráfico de barra Horizontal
sns.barplot(x='Preços',y='Produtos',data=produtos,hue='Ano',palette=cores2,orient='h')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Quero saber o total de vendas separado por categoria, de todas as categorias, de acordo com o seu método de pagamento

pedidos = pd.read_csv('./pedidos.csv')
# Total de vendas para cada pedido
pedidos['total_venda'] = pedidos['quantidade'] * pedidos['preco_unitario']
# Agrupar total de vendas por categoria
vendas_por_categoria = pedidos.groupby(['categoria','metodo_pagamento']).agg({'total_venda':'sum'}).sort_values('total_venda').reset_index()

plt.figure(figsize=(10,6),dpi=150)
sns.barplot(x='categoria',y='total_venda',data=vendas_por_categoria,hue='metodo_pagamento',palette='bright')
plt.legend(title='Legenda',loc='lower right')
plt.show()

plt.figure(figsize=(10,6),dpi=150)
sns.barplot(x='total_venda',y='categoria',data=vendas_por_categoria,hue='metodo_pagamento',palette='bright',orient='h')
plt.legend(title='Legenda',loc='lower left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Quero saber o total de vendas separado por categoria, de todas as categorias, de acordo com o seu método de pagamento

pedidos = pd.read_csv('./pedidos.csv')
# Total de vendas para cada pedido
pedidos['total_venda'] = pedidos['quantidade'] * pedidos['preco_unitario']
# Filtrar por categoria Lazer e Eletrônicos
dados_agrupados = pedidos[pedidos['categoria'].isin(['Lazer','Eletrônicos'])]
# Agrupar total de vendas por categoria
vendas_por_categoria = dados_agrupados.groupby(['categoria','metodo_pagamento']).agg({'total_venda':'sum'}).sort_values('total_venda').reset_index()

plt.figure(figsize=(10,6),dpi=150)
sns.barplot(x='categoria',y='total_venda',data=vendas_por_categoria,hue='metodo_pagamento',palette='bright')
plt.legend(title='Legenda',loc='lower right')
plt.show()

plt.figure(figsize=(10,6),dpi=150)
sns.barplot(x='total_venda',y='categoria',data=vendas_por_categoria,hue='metodo_pagamento',palette='bright',orient='h')
plt.legend(title='Legenda',loc='lower left')
plt.show()
```
