```python
# Desafio 1

# Crie um gráfico de barra com base no arquivo cursos.csv(em anexo) que mostra a média da mensalidade de todos os cursos(algo que precisa ser calculado), onde a legenda mostra o ano que está sendo representado por cada barra de cada curso

# Seu Gráfico de barra deve conter um tamanho adequado o suficiente para que os dados fiquem claramente visíveis e estar ordenado por mensalidade.
# Utilizar a paleta de cores 'bright'

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados de vendas
cursos = pd.read_csv('./cursos.csv')

cursos['Ano'] = pd.to_datetime(cursos['Data']).dt.year

cursos_agrupados = cursos.groupby(['Curso','Ano']).agg({'Mensalidade':'mean'}).sort_values('Mensalidade').reset_index()


plt.figure(figsize=(20,5),dpi=150)
sns.barplot(x='Curso',y='Mensalidade',data=cursos_agrupados,hue='Ano',palette='bright')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados de vendas
cursos = pd.read_csv('./cursos.csv')

cursos['Ano'] = pd.to_datetime(cursos['Data']).dt.year

cursos_agrupados = cursos.groupby(['Curso','Ano']).agg({'Mensalidade':'mean'}).sort_values('Mensalidade').reset_index()
cursos_agrupados.head(20)

# Desafio 2 - Exibir os resultados do desafio 1 na horizontal
plt.figure(figsize=(10,10),dpi=150)
sns.barplot(x='Mensalidade',y='Curso',data=cursos_agrupados,hue='Ano',palette='bright',orient='h')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados de vendas
cursos = pd.read_csv('./cursos.csv')

cursos['Ano'] = pd.to_datetime(cursos['Data']).dt.year

cursos_filtrados = cursos[cursos['Ano'].isin([2013,2014,2015]) & (cursos['Curso'].isin(['Engenharia Civil','Medicina']))]

cursos_agrupados = cursos_filtrados.groupby(['Curso','Ano']).agg({'Mensalidade':'mean'}).sort_values('Mensalidade').reset_index()
cursos_agrupados.head(20)

plt.figure(figsize=(10,10),dpi=150)
sns.barplot(x='Curso',y='Mensalidade',data=cursos_agrupados,hue='Ano',palette='bright')
plt.show()
```
