```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

vendas_2022 = pd.read_csv('./2022_data.csv')
vendas_2023 = pd.read_csv('./2023_data.csv')

vendas_2022['Ano'] = '2022'
vendas_2023['Ano'] = '2023'

dados_combinados = pd.concat([vendas_2022,vendas_2023])

plt.figure(figsize=(12,6),dpi=150)
sns.histplot(data=dados_combinados,x='metodo_pagamento',hue='Ano',multiple='stack',palette=['purple','yellow']) #multiple=dodge
plt.show()
```
