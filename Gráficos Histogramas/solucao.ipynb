```python
# Desafio 4

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

cursos2022 = pd.read_csv('./cursos_2022.csv')
cursos2023 = pd.read_csv('./cursos_2023.csv')

cursos2022['Ano'] = '2022'
cursos2023['Ano'] = '2023'

cursos_agrupados = pd.concat([cursos2022,cursos2023])

plt.figure(figsize=(10,5),dpi=150)
sns.histplot(data=cursos_agrupados,x='Turno',kde=True,hue='Ano',palette=['red','gold'],multiple='dodge')
plt.show()
```
