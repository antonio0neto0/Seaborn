Tipos de Gráficos: Gráficos de Pizza (Pie Chart)

```python
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm # código novo

for font in fm.fontManager.ttflist:
    print(font.name)

sns.set_theme(palette='bright',font='DejaVu Sans Mono',font_scale=1.5) # Novo linha do Seaborn

categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [10, 20, 30, 40]

plt.figure(figsize=(7, 7))
explode = (0.1, 0, 0, 0)  
plt.pie(valores, labels=categorias, autopct='%1.1f%%', explode=explode, startangle=90, textprops={'color': 'black'})
plt.title('Distribuição por Categorias')
plt.show()

```
