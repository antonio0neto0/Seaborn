```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dados_faculdade = pd.read_csv('./cursos.csv')

dados_faculdade['Data'] = pd.to_datetime(dados_faculdade['Data'])

dados_faculdade['Ano'] = dados_faculdade['Data'].dt.year

cursos_noturnos = dados_faculdade[(dados_faculdade['Turno'] == 'Noturno') & (dados_faculdade['Ano'].isin([2018,2019,2020,2021,2022])) & (dados_faculdade['Curso'].isin(['Medicina','Odontologia','Enfermagem']))]

cursos_agrupados = cursos_noturnos.groupby(['Curso','Ano'])['Mensalidade'].mean().reset_index(name='Média Mensalidade')

sns.set_theme(style='darkgrid')
plt.figure(figsize=(13.5,5),dpi=150)
sns.lineplot(data=cursos_agrupados,x='Ano',y='Média Mensalidade',hue='Curso',style='Curso',palette=['green','purple','orange'],markers=True,dashes=True)

plt.title('Média das Mensalidades entre 2018 e 2022',fontsize=16)
plt.xlabel('Ano',fontsize=14)
plt.ylabel('Média de Mensalidade',fontsize=14)
plt.show()
```
