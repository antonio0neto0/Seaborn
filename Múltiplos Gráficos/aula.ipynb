```python
# Para criar Dashboards -> Subplots
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Subplots
figure, axes  =plt.subplots(nrows=2,ncols=2,figsize=(30,10))
plt.tight_layout(pad=3)
# 1 linha, 2 colunas
##

# 2 linhas, 2 colunas
## 
## 

# 3 linha, 2 colunas
##
##
##

# Preparação gráfico 1
cidades = ['São Paulo', 'Rio de Janeiro', 'Brasília', 'Salvador', 'Fortaleza']
populacao = [12180000, 6710000, 3050000, 2886000, 2800000]

# Plotar gráfico 1
sns.barplot(x=cidades, y=populacao,hue=cidades, palette='bright',ax=axes[0,0])
axes[0,0].set_title('População das Maiores Cidades do Brasil')
axes[0,0].set_xlabel('Cidade')
axes[0,0].set_ylabel('População(em milhões)')

# Preparação gráfico 2
idade_populacao = [22, 25, 30, 35, 40, 42, 45, 50, 55, 60, 65, 70, 75, 80, 85]
# Plotar gráfico 2
sns.histplot(idade_populacao, bins=5, kde=False, edgecolor='black', color='skyblue',ax=axes[0,1])
axes[0,1].set_title('Distribuição da Idade da População')
axes[0,1].set_xlabel('Idade')
axes[0,1].set_ylabel('Frequência')


# Preparação gráfico 3
altura = np.random.normal(170, 10, 100)
peso = altura * 0.7 + np.random.normal(0, 5, 100)
# Plotar gráfico 3
sns.scatterplot(x=altura, y=peso, color='blue', alpha=0.5,ax=axes[1,0])
axes[1,0].set_title('Altura vs. Peso')
axes[1,0].set_xlabel('Altura (cm)')
axes[1,0].set_ylabel('Peso (kg)')


# Preparação gráfico 4
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
vendas = [150, 200, 170, 180, 160, 175]
# Plotar gráfico 3
sns.lineplot(x=meses, y=vendas, marker='o', linestyle='--', color='green', label='Vendas Mensais',ax=axes[1,1])
axes[1,1].set_title('Vendas Mensais no Primeiro Semestre')
axes[1,1].set_xlabel('Meses')
axes[1,1].set_ylabel('Vendas')
axes[1,1].legend(title='Legenda', loc='upper right')

plt.show()


```
