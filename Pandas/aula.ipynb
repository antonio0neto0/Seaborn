```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

pedidos = pd.read_csv('./vendas_instagram.csv')
# Agrupar produtos pelo seu total de vendas
total_pedidos = pedidos.groupby('produto')['preco'].sum().reset_index()
total_pedidos.head(5)

plt.figure(figsize=(15,5),dpi=150)
sns.lineplot(x=total_pedidos['produto'],y=total_pedidos['preco'])
plt.show()

```
