```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grfico de Linha - Seaborn
meses = ['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio', 'Junho']
vendas = [150, 200, 170, 180, 160, 175]
# Tema do GRID(grade)
sns.set_theme(style='darkgrid') # darkgrid, whitegrid, dark, white
plt.figure(figsize=(10,6),dpi=150,facecolor='lightblue',frameon=True)
# solid (padro),dashed, dotted,dashdot
sns.lineplot(x=meses,y=vendas,marker='o',color='red',linestyle='dashdot',label='Vendas')
plt.legend(title='Legenda',loc='lower right') # upper right, upper left, lower right, lower left
plt.title('Vendas Mensais', fontsize=16)
plt.ylabel('Vendas',fontsize=14)
plt.xlabel('Meses',fontsize=14)
plt.show()
```
