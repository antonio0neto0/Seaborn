```python
import seaborn as sns
import matplotlib.pyplot as plt

# Desafio: Personalizao 
# Com base nos dados fornecidos, crie um grfico que mostre a evoluo das vendas de acordo com os meses do segundo semestre (ou seja, dos meses fornecidos aqui).

# Instrues do Desafio:
# - Estilo de linha: slida
# - Cor da linha: #1efac3
# - Texto da legenda: Legenda
# - Posio da legenda: canto inferior direito
# - Legenda eixo X: Meses
# - Legenda eixo Y: Vendas
# - Grid: ligado - darkgrid
# - Tamanho do grfico: 10x5 (polegadas)
# - DPI: 150
# - Cor externa: #1efac3

# Dados:
meses_desafio = ['Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas_desafio = [190, 210, 180, 200, 195, 220]

# Configuraes Iniciais
sns.set_theme(style='darkgrid') 

# Criando e Personalizando o Grfico
plt.figure(figsize=(10, 5), facecolor='#1efac3', dpi=150, frameon=True) # Configuramos o tamanho, cor de fundo e dpi do grfico
sns.lineplot(x=meses_desafio, y=vendas_desafio, color='#1efac3', marker='s', linestyle='solid', label='Vendas') # Usamos listas diretamente

# Personalizaes Adicionais
plt.title('Evoluo das Vendas no Segundo Semestre', fontsize=16) # Adicionamos um ttulo ao grfico
plt.xlabel('Meses', fontsize=14) # Definimos a legenda do eixo X
plt.ylabel('Vendas', fontsize=14) # Definimos a legenda do eixo Y
plt.legend(title='Legenda', loc='lower right') # Configuramos a legenda conforme solicitado

plt.show()
```
