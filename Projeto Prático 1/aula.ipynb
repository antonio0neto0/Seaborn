Tarefa 1: An√°lise de Performance por Categoria em 2024

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando o arquivo CSV
dados_vendas = pd.read_csv('./Vendas_de_Produtos.csv')

# Convers√£o para o tipo apropriado para manipula√ß√£o de datas
dados_vendas['Data da Venda'] = pd.to_datetime(dados_vendas['Data da Venda'])

# Filtragem por vendas de 2024
dados_2024 = dados_vendas[dados_vendas['Data da Venda'].dt.year == 2024]

# Agrupamento de dados para plotagem
vendas_por_categoria_2024 = dados_2024.groupby('Produto')['Total da Venda'].sum().reset_index()

# Cria√ß√£o do gr√°fico de barra
sns.set_style(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Produto', y='Total da Venda', hue='Produto',data=vendas_por_categoria_2024, palette='viridis')
plt.title('Total de Vendas por Categoria de Produto em 2024',fontsize=16)
plt.xlabel('Categoria de Produto',fontsize=14)
plt.ylabel('Total de Vendas ($)',fontsize=14)
plt.show()
```

 Tarefa 2: Tend√™ncias de Vendas Mensais

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Leitura do arquivo
vendas = pd.read_csv('./Vendas_de_Produtos.csv')

# Convers√£o para o tipo apropriado para manipula√ß√£o de datas
vendas['Data da Venda'] = pd.to_datetime(vendas['Data da Venda'])

# Extra√ß√£o e cria√ß√£o da coluna m√™s
vendas['M√™s'] = vendas['Data da Venda'].dt.month

# Filtragem por vendas de 2022
vendas_filtradas = vendas[vendas['Data da Venda'].dt.year == 2022]

# Agrupamento de dados para plotagem
total_de_vendas = vendas_filtradas.groupby('M√™s')['Total da Venda'].mean().reset_index()

# Cria√ß√£o do gr√°fico de linha
plt.figure(figsize=(10,5),dpi=150)
sns.set_theme(style=('darkgrid'))
sns.lineplot(data=total_de_vendas,x='M√™s',y='Total da Venda',marker='o',color='blue')
plt.title('M√©dia de Vendas por m√™s')
plt.show()
```

Tarefa 3: Impacto dos Finais de Semana nas Vendas

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Cria√ß√£o de fun√ß√£o que categoriza os dias
def eh_fim_de_semana(data):
    return 'Fim de Semana' if data in (5,6) else 'Dia √∫til'

# Leitura do arquivo
vendas = pd.read_csv('./Vendas_de_Produtos.csv')

# Convers√£o para o tipo apropriado para manipula√ß√£o de datas
vendas['Data da Venda'] = pd.to_datetime(vendas['Data da Venda'])

# Filtragem por vendas de 2023
vendas_filtradas = vendas[vendas['Data da Venda'].dt.year == 2023]

# Utiliza√ß√£o da fun√ß√£o eh_fim_de_semana para categorizar cada dia da semana
vendas['Fim de Semana'] = vendas['Data da Venda'].dt.dayofweek.apply(eh_fim_de_semana)

# Agrupamento de dados para plotagem
vendas_agrupadas = vendas.groupby('Fim de Semana')['Total da Venda'].sum().reset_index()

# Cria√ß√£o de gr√°fico de barra
plt.figure(figsize=(10,5),dpi=150)
sns.set_theme(style=('darkgrid'))
sns.barplot(data=vendas,x='Fim de Semana',y='Total da Venda',hue='Fim de Semana',palette=['green','blue'])
plt.title('M√©dia de Vendas por m√™s',fontsize=16)
plt.xlabel('Vendas por dia da Semana',fontsize=14)
plt.ylabel('Total de Vendas',fontsize=14)
plt.show()
```
