```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dados
meses = [1, 2, 3, 4, 5, 6]
vendas = [190, 210, 180, 200, 195, 220]

# Configurando o tema
sns.set_theme(style='darkgrid')

# Criando o gráfico de linha com Seaborn
plt.figure(figsize=(10, 6),dpi=150) 
sns.lineplot(x=meses, y=vendas)

# Adicionando título e legendas aos eixos
plt.title('Vendas Mensais', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)

plt.text(x=2.78,y=178.5,s='Pior Mês',color='red')
plt.text(x=5.7,y=221,s='Melhor Mês',color='green',backgroundcolor='black')

# Adicionando uma anotação com seta
# xy - coordenadas de onde finaliza(ou para onde está apontando a seta)
# xytext - coordenadas de onde inicia a seta e o texto
# 'arrowprops' define as propriedades da seta.
plt.annotate(text='Média',xytext=(3.68,205),xy=(4,199.92),arrowprops=dict(color='red',arrowstyle='->'))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dados
meses = ['Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas = [190, 210, 180, 200, 195, 220]

# Configurando o tema
sns.set_theme(style='darkgrid')

# Criando o gráfico de linha com Seaborn
plt.figure(figsize=(10, 6),dpi=150) 
sns.lineplot(x=meses, y=vendas)

# Adicionando título e legendas aos eixos
plt.title('Vendas Mensais', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)

# Como adicionar texto usando Matplotlib
plt.text(meses.index('Setembro') -0.2, 178.5, "Pior Mês", fontsize=10, color='red')
plt.text(meses.index('Dezembro') -0.3, 220, "Melhor Mês", fontsize=10, color='green', backgroundcolor='black')

# Exibindo o gráfico
plt.show()

# Neste exemplo, usamos Seaborn para criar o gráfico de linha por ser mais direto e oferecer uma estética melhorada com menos código.
# Após criar o gráfico com Seaborn, usamos funções do Matplotlib para adicionar textos personalizados ao gráfico,
# permitindo destacar informações específicas como o pior e o melhor mês em termos de vendas.
```

```python
import matplotlib.pyplot as plt

# Define some data points
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create subplots for each arrow style
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# Arrow Style: '->'
ax1 = axes[0, 0]
ax1.plot(x, y, 'bo')
ax1.annotate('Arrow Style: ->', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='->',color='red'))
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Arrow Style: ->')

# Arrow Style: '-['
ax2 = axes[0, 1]
ax2.plot(x, y, 'bo')
ax2.annotate('Arrow Style: -[', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='-[',color='red'))
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Arrow Style: -[')

# Arrow Style: '|-|'
ax3 = axes[0, 2]
ax3.plot(x, y, 'bo')
ax3.annotate('Arrow Style: |-|', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='|-|',color='red'))
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_title('Arrow Style: |-|')

# Arrow Style: '-|>'
ax4 = axes[1, 0]
ax4.plot(x, y, 'bo')
ax4.annotate('Arrow Style: -|>', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='-|>',color='red'))
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_title('Arrow Style: -|>')

# Arrow Style: '<-'
ax5 = axes[1, 1]
ax5.plot(x, y, 'bo')
ax5.annotate('Arrow Style: <-', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='<-',color='red'))
ax5.set_xlabel('X')
ax5.set_ylabel('Y')
ax5.set_title('Arrow Style: <-')

# Arrow Style: '<->'
ax6 = axes[1, 2]
ax6.plot(x, y, 'bo')
ax6.annotate('Arrow Style: <->', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='<->',color='red'))
ax6.set_xlabel('X')
ax6.set_ylabel('Y')
ax6.set_title('Arrow Style: <->')

# Arrow Style: 'fancy'
ax7 = axes[2, 0]
ax7.plot(x, y, 'bo')
ax7.annotate('Arrow Style: fancy', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='fancy',color='red'))
ax7.set_xlabel('X')
ax7.set_ylabel('Y')
ax7.set_title('Arrow Style: fancy')

# Arrow Style: 'simple'
ax8 = axes[2, 1]
ax8.plot(x, y, 'bo')
ax8.annotate('Arrow Style: simple', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='simple',color='red'))
ax8.set_xlabel('X')
ax8.set_ylabel('Y')
ax8.set_title('Arrow Style: simple')

# Arrow Style: 'wedge'
ax9 = axes[2, 2]
ax9.plot(x, y, 'bo')
ax9.annotate('Arrow Style: wedge', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='wedge',color='red'))
ax9.set_xlabel('X')
ax9.set_ylabel('Y')
ax9.set_title('Arrow Style: wedge')

plt.tight_layout()
plt.show()
```
