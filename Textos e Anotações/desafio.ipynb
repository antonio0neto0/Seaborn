```python
# Dado o código a seguir, insira texto de acordo com os ponto mencionados nessa imagem: https://i.imgur.com/RqhBsMH.png
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando os dados
meses_desafio = [10, 20, 30, 40, 50, 60]
vendas_desafio = [190, 210, 180, 200, 195, 220]

plt.figure(figsize=(10, 6))  # Define o tamanho do gráfico
sns.lineplot(x=meses_desafio, y=vendas_desafio, linestyle='-', color='blue')

plt.title('Vendas Mensais', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)

# Insira o texto "2° maior mês" no ponto 1, com texto preto e fundo verde
# Insira o texto "3° Maior mês" no ponto 2, com texto verde e fundo amarelo(yellow)
plt.text(16, 211, "2° maior mês", fontsize=10, color='black', backgroundcolor='green')
plt.text(45, 192, "4° maior mês", fontsize=10, color='green', backgroundcolor='yellow')

plt.show()
```

```python
# Dado o código a seguir, insira textos, de acordo com a imagem a seguir: https://i.imgur.com/Xexlttc.png
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando os dados
meses_desafio = ['Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas_desafio = [190, 210, 180, 200, 195, 220]

plt.figure(figsize=(10, 6))  # Ajusta o tamanho do gráfico
sns.lineplot(x=meses_desafio, y=vendas_desafio, marker='o', linestyle='-', color='blue', label='Vendas Mensais')

plt.title('Vendas Mensais', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)

# Insira o texto "início" no ponto 1, com texto azul claro(lightblue)e fundo preto
# Insira o texto "média" no ponto 2, com texto amarelo(yellow) e fundo preto
plt.text(meses_desafio.index("Julho") - 0.2, 188, "Início", fontsize=10, color='lightblue', backgroundcolor='black')
plt.text(meses_desafio.index("Outubro") - 0.2, 202, "Média", fontsize=10, color='yellow', backgroundcolor='black')

plt.show()
```

```python
# Dado o código a seguir, crie duas setas similar ao que vê na imagem a seguir: https://i.imgur.com/uaNFt6l.png
import seaborn as sns
import matplotlib.pyplot as plt

meses_desafio = [10, 20, 30, 40, 50, 60]
vendas_desafio = [190, 210, 180, 200, 195, 220]

sns.lineplot(x=meses_desafio, y=vendas_desafio, marker='o', linestyle='-', color='blue')

plt.title('Vendas Mensais', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)
plt.annotate(text='ponto 1', xy=(20, 210), xytext=(20, 220),arrowprops=dict(color='red', arrowstyle='->'))
plt.annotate(text='ponto 2', xy=(40, 200), xytext=(45, 215),arrowprops=dict(color='red', arrowstyle='->'))

plt.show()
```
