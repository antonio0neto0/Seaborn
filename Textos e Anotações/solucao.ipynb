```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dados
meses = [10, 20, 30, 40, 50, 60]
vendas = [190, 210, 180, 200, 195, 220]

# Configurando o tema
sns.set_theme()

# Criando o grfico de linha com Seaborn
plt.figure(figsize=(10, 6)) # Define o tamanho do grfico
sns.lineplot(x=meses, y=vendas)

# Adicionando ttulo e legendas aos eixos
plt.title('Vendas Mensais', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)

# Como adicionar texto usando Matplotlib
plt.text(28, 178.5, "Pior Ms", fontsize=10, color='red')
plt.text(57, 220, "Melhor Ms", fontsize=10, color='green', backgroundcolor='black')

# Exibindo o grfico
plt.show()

# Neste exemplo, usamos Seaborn para criar o grfico de linha por ser mais direto e oferecer uma esttica melhorada com menos cdigo.
# Aps criar o grfico com Seaborn, usamos funes do Matplotlib para adicionar textos personalizados ao grfico,
# permitindo destacar informaes especficas como o pior e o melhor ms em termos de vendas.
```
